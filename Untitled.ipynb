```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("alb_homes.csv")
```

```python
df.head(3)
```

```python
df.lotsize.describe()
```

```python
df.cooling = df.cooling.apply((lambda x: 0 if x == 'No Central Air' else 1), convert_dtype=True)
df.condition = pd.Categorical(df.condition)
df.esdistrict = pd.Categorical(df.esdistrict)
df.msdistrict = pd.Categorical(df.msdistrict)
df.hsdistrict = pd.Categorical(df.hsdistrict).codes
```

```python
df.drop(columns=['yearbuilt'], inplace=True)
```

```python
df.corr().totalvalue
```

```python
sns.pairplot(data = df, y_vars = ['totalvalue'])
```

```python
sns.boxplot(y=np.log(df.totalvalue), x=df.esdistrict)
```

```python
sns.scatterplot(y=np.log(df.totalvalue), x=df.finsqft)
```

```python
sns.histplot(x=np.log(df.totalvalue), hue=df.finsqft)
```
